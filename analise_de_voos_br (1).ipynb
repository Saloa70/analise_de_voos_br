## Explorando o DataSet

### Quantidade de Voos por Ano e Mês

```python
df.info()
```

```python
df.groupby(df['dt.Part_Prev'].dt.year).agg({'Voos': np.count_nonzero}).reset_index()
```

```python
sns.set_style("darkgrid")

ax = sns.barplot(
    data = df.groupby(df['dt.Part_Prev'].dt.year).agg({'Voos': np.count_nonzero}).reset_index(),
    x = 'dt.Part_Prev',
    y = 'Voos',
    alpha= .8
    )

ax.figure.set_size_inches(12,6)
ax.set_title("Quantidade de Voos x Ano - Data Prevista de Partida")
ax.set_xlabel("Ano")
ax.set_ylabel("Quantidade de Voos")

#plt.tickLabel_format(style='plain', axis='y')

ax
```

```python
df.assign(ano=df['dt.Part_Prev'].dt.year, mes=df['dt.Part_Prev'].dt.month
         ).groupby(['ano','mes']).agg({'Voos': np.count_nonzero}).reset_index().head(5)
```

```python
ax = sns.barplot(
    data = df.assign(ano=df['dt.Part_Prev'].dt.year, mes=df['dt.Part_Prev'].dt.month
         ).groupby(['ano','mes']).agg({'Voos': np.count_nonzero}).reset_index(),
    x = 'mes',
    hue = 'ano',
    y = 'Voos',
    alpha = .8
)

ax.figure.set_size_inches(18,6)
ax.set_title("Quantidade de Voos x Ano - Data Prevista de Partida")
ax.set_xlabel("Ano")
ax.set_ylabel("Quantidade de Voos")

#plt.legend(bbox_to_anchor= (1.0, 0.6))

ax


```

```python
df.assign(
    ano=df['dt.Part_Prev'].dt.year,
    mes=df['dt.Part_Prev'].dt.strftime('%b'),
    mes_ordem = df['dt.Part_Prev'].dt.month
         ).groupby(['ano', 'mes', 'mes_ordem']).agg({'Voos': np.count_nonzero}).reset_index().sort_values(by='mes_ordem')

ax = sns.barplot(
    data = df.assign(
    ano=df['dt.Part_Prev'].dt.year,
    mes=df['dt.Part_Prev'].dt.strftime('%b'),
    mes_ordem = df['dt.Part_Prev'].dt.month
         ).groupby(['ano', 'mes', 'mes_ordem']).agg({'Voos': np.count_nonzero}).reset_index().sort_values(by='mes_ordem'),
    x = 'mes',
    hue = 'ano',
    y = 'Voos',
    alpha = .8
)

ax.figure.set_size_inches(18,6)
ax.set_title("Quantidade de Voos x Ano - Data Prevista de Partida")
ax.set_xlabel("Ano")
ax.set_ylabel("Quantidade de Voos")


ax
```

```python
df.groupby('Aeroporto_Origem').agg({'Voos': np.count_nonzero}).reset_index().sort_values(by='Voos', ascending=False)
```

```python
df_origem = df.groupby('Aeroporto_Origem').agg({'Voos': np.count_nonzero}).reset_index().sort_values(by='Voos', ascending=False)

df_origem.columns = ['Origem', 'Qtde de Voos']

df_origem['Perc_Voos'] = ((df_origem['Qtde de Voos'] / df.shape[0] ) * 100 ).round(3)

df_origem['Acumulado'] = df_origem['Perc_Voos'].cumsum()

df_origem.head(20)
```

```python
ax = sns.barplot(
    data = df_origem.head(20),
    x = 'Qtde de Voos',
    y = 'Origem'
)

ax.figure.set_size_inches(6,8)
ax.set_title('Aeroporto de Origem x Quantidade de Voos')
ax.set_xlabel('Quantidade de Voos')
ax.set_ylabel('')
```

## Qual o Tipo de Voo mais Utilizado ?

```python
df['Tipo_Voo'].value_counts()
```

```python
df.assign(
    dt=df['dt.Part_Prev'].dt.strftime('%Y-%m-%d').astype('datetime64[ns]'),
    tp=df['Tipo_Voo']).groupby(['dt', 'tp']).agg({'Voos': np.count_nonzero}).reset_index()
           
ax = sns.lineplot(
    data = df.assign(
        dt =df['dt.Part_Prev'].dt.strftime('%Y-%m-%d').astype('datetime64[ns]'),
        tp=df['Tipo_Voo']).groupby(['dt', 'tp']).agg({'Voos': np.count_nonzero}).reset_index(),
    x = 'dt',
    y = 'Voos',
    hue = 'tp',
    marker = 'o'
)
ax.figure.set_size_inches(15,5)
           
```

```python
import matplotlib.pyplot as plt

for tp_nome in ['Internacional', 'Nacional', 'Regional']:

    ax = sns.lineplot(
        data = df[ df['Tipo_Voo'] == tp_nome ].assign(
            dt=df['dt.Part_Prev'].dt.strftime('%Y-%m-%d').astype('datetime64[ns]'),
            tp=df['Tipo_Voo']).groupby(['dt', 'tp']).agg({'Voos': np.count_nonzero}).reset_index(),
        x = 'dt',
        y = 'Voos',
        hue = 'tp',
        marker = 'o'
    )
    
    ax.figure.set_size_inches(15,5)
    ax.set_title(tp_nome.upper())
    plt.figure()
    
```
